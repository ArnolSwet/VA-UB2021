a) Each of the scenes in a video is usually called 'shot'. Find where a shot (scene) finishes and the following starts (the shot boundaries). 

To solve this exercise, you need to create a **temporal plot** showing a frame by frame difference measure to be defined by you, being applicable to distinguish the shots. Define a criterion to detect the boundaries of the shots and visualize it in a static plot.

```python
from skimage import img_as_float
from skimage.exposure import histogram

def compute_hist(im, im_prev, hist_im_prev=None):
    """
    Function for computing the difference of histogram of an image and the previous one.
    
    :param im: the current image of the video
    :param im_prev: the previous image of the video
    :hist_im_prev (optional): the histogram of the previous image
    :return: the difference of histogram of the image and the previous one and the histogram of the current image
    """
    #Compute the histogram of each channel of the current image with 32 bins
    #We have considered to use 32 bins because it is enough to detect the changes of scenes
    [hist_r, bins_r] = histogram((img_as_float(im)[:,:,0]),32)
    [hist_g, bins_g] = histogram((img_as_float(im)[:,:,1]),32)
    [hist_b, bins_b] = histogram((img_as_float(im)[:,:,2]),32)
    #Finally, we sum the three histograms to obtain a single histogram
    hist_im = np.array(hist_r + hist_g + hist_b)
    
    if hist_im_prev is None: #If the parameter hist_im_prev is None, we have to compute the histogram of the previous image
        #Compute the histogram of each channel of the previous image with 32 bins
        [hist_r, bins_r] = histogram((img_as_float(im_prev)[:,:,0]),32)
        [hist_g, bins_g] = histogram((img_as_float(im_prev)[:,:,1]),32)
        [hist_b, bins_b] = histogram((img_as_float(im_prev)[:,:,2]),32)
        #Finally, we sum the three histograms to obtain a single histogram
        hist_im_prev = np.array(hist_r + hist_g + hist_b)
        
    #Calculate the difference between the two histograms in absolute value and take the mean. This is the measure
    #we have defined for each image. It will help us to detect if there has been a change of scene.
    hist_diff = np.mean(np.abs(hist_im_prev - hist_im))
    
    #Return the difference between the two histograms and the histogram of the current image.
    return hist_diff, hist_im



#Now, let's visualize a graphic showing the frame by frame measure that we have defined and then,
#looking at the result, we will be able to define a criterion to detect the boundaries of the shots.

#Initialize the data arrays 
xdata, ydata = [], []

%matplotlib inline
    #Changing the pluggin to visualize the graphic

prev_hist = None #We use this variable to use it as parameter for the function compute_hist to compute the histogram of
                 #each image only once.

for t in range(len(ic)-1):
    #Compute the measure for the shot number 't+1'
    y, prev_hist = compute_hist(ic[t+1], ic[t], hist_im_prev=prev_hist)
    xdata.append(t) # x axis = time
    ydata.append(y) # y axis = measure
    
fig, ax = plt.subplots(1)
ax.set_title("Difference between the histogram of an image and the previous image's one")
ax.set_xlabel('Index image in the video')
ax.set_ylabel('Difference')
ax.plot(xdata, ydata)

plt.show()
```

## 3.2 Background subtraction

Apply the background subtraction algorithm (check theory material).

a) Visualize the following images for each different scene (there are 3) of the video:

    1) an image belonging to the shot
    2) the background image, and
    3) the foreground.
    
**Hint**: You can construct a mask obtained from the original image and the background in order to know which parts of the image form part from the foreground and recover from the original image just the foreground regions.

```python
# We simply plot the results obtained previously.
f, axarr = plt.subplots(ncols=2, nrows=3,figsize=(8,8))
for i in range(len(shot_beginnings)-1):
    axarr[i,0].imshow(ic[shot_beginnings[i]])
    axarr[i,0].set_title('Shot ' + str(i+1) + ' beginning') # We set the title
    axarr[i,0].axis('off')
    axarr[i,1].imshow(ic[shot_beginnings[i+1]-1])
    axarr[i,1].set_title('Shot ' + str(i+1) + ' ending') # We set the title
    axarr[i,1].axis('off')
plt.show()

# Function that, for each pixel, computes the median between the whole set of images. That will be the background.
def get_background(shot_frames):
    background = np.zeros(shot_frames[0].shape + (len(shot_frames),)) #all images have the same shape, we add one dimension
    
    for j in range(len(shot_frames)): background[:,:,:,j] = shot_frames[j]

    background = np.median(background, axis = 3)
    return background.astype(np.int)

# We plot the results
shape = ic[0].shape 
f, axarr = plt.subplots(ncols=1, nrows=3,figsize=(8,8))
results = []
for i in range(len(shot_beginnings)-1):
    result = get_background(ic[shot_beginnings[i] : shot_beginnings[i+1]])
    results += [result]
    
    axarr[i].imshow(result)
    axarr[i].set_title('Shot ' + str(i+1) + ' background') # We set the title
    axarr[i].axis('off')
    
plt.show()

from skimage import color

# The foreground is the result of substracting the background to the image
def get_foreground(background, image):
    result = image - background
    result[result < 0] = 0
    return result;

# We show the foreground for the beginning and ending of every shot.
f, axarr = plt.subplots(ncols=2, nrows=3,figsize=(8,8))
threshold = 100
for i in range(len(shot_beginnings)-1):
    axarr[i,0].imshow(get_foreground(results[i], ic[shot_beginnings[i]]))
    axarr[i,0].set_title('Shot ' + str(i+1) + ' foreground') # We set the title
    axarr[i,0].axis('off')
    axarr[i,1].imshow(get_foreground(results[i], ic[shot_beginnings[i+1]-1]))
    axarr[i,1].set_title('Shot ' + str(i+1) + ' foreground') # We set the title
    axarr[i,1].axis('off')
plt.show()
```
